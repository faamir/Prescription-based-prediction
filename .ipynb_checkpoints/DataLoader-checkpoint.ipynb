```python
df_tokenized = df_labeled.values[:,0:-1]
df_tokenized = df_tokenized.astype(dtype=str)
labels = df_labeled.values[:,-1]
df_tokenized
```

```python
# build vocabulary, word2index and index2word
def vocab_builder(sequences):
    word_cnt = Counter(itertools.chain(*sequences))
    vocab_invariant = [x[0] for x in word_cnt.most_common()]
    vocabs = {x: i for i, x in enumerate(vocab_invariant)}
    # print("vocabs", vocabs)
    # print("vocab_invariant", vocab_invariant)
    return [vocabs, vocab_invariant]

# map sentences
def mapper(sequences, labels, vocabs):
    # ye list besaz ba tavajo be index gozari tu tabe ghabli baraye har jomle
    x = np.array([[vocabs[str(word)] for word in sentence] for sentence in sequences])
    y = np.array(labels)
    return [x, y]
```

```python
vocabs, vocab_invariant = vocab_builder(df_tokenized)
x, y = mapper(df_tokenized, labels, vocabs)
```

```python
#inputs and lables
print(x)
print(y)
```

```python
#Normalize inputs between 0 and 1
#x= keras.preprocessing.sequence.pad_sequences(x, maxlen=20, truncating='post', padding='post')
min_max_scaler = preprocessing.MinMaxScaler()
x = min_max_scaler.fit_transform(x)
```

```python
#Shuffle and Split input data
np.random.seed(42)
shuffling = np.random.permutation(np.arange(len(y)))
x = x[shuffling]
y = y[shuffling]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
print("Number of train data: ", X_train.shape)
print("Number of test data: ", X_test.shape)
```
